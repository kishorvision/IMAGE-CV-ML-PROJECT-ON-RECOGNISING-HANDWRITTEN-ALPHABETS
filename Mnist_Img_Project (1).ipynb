# MAKING DATAFRAME

```python
import os
```

```python
from matplotlib import pyplot as plt
from matplotlib import image
from PIL import Image
```

```python
label=[]
target=[]
file="B:/Ext_img/"
os.listdir(file)
for files in os.listdir(file):
    for file_1 in os.listdir(os.path.join(file,files)):
        file_2=os.path.join(os.path.join(file,files),file_1)
        img=Image.open(file_2)
        img1=img.resize((8,8))
        img2=np.array(img1).ravel()
        label.append(img2)
        target.append(files)
```

```python
file="B:/Ext_img/"
```

```python
os.listdir(file)
```

```python
df=pd.DataFrame(label,columns=["pixel"+ str(i) for i in range(0,64)])
df["label"]=target
```

# BUILDING THE MODEL

## `Logistic Regression`

```python
%%time
from sklearn.linear_model import LogisticRegression
lrc=LogisticRegression()
lrc.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score, recall_score,classification_report
```

```python
%%time
y_pred_lrc=lrc.predict(X_test)
y_pred_lrc
```

```python
accuracy_score(y_test,y_pred_lrc)
```

```python
confusion_matrix(y_test,y_pred_lrc)
```

```python
temp_df=pd.DataFrame({"Actual":y_test,"Predicted":y_pred_lrc})
temp_df.head(10)
```

```python
temp_df['Correctly Predicted'] = temp_df['Actual'] == temp_df['Predicted']
```

```python
correctly_predicted_df = temp_df[temp_df['Correctly Predicted']]
wrongly_predicted_df = temp_df[~temp_df['Correctly Predicted']]
```

```python
print("Correctly Predicted Values:")
print(correctly_predicted_df)

print("\nWrongly Predicted Values:")
print(wrongly_predicted_df)
```

```python
pd.crosstab(y_test, y_pred_lrc, rownames=['Actual'], colnames=['Predicted'])
```

```python
sns.histplot(y_test,color='blue',alpha=0.5)
sns.histplot(y_pred_lrc,color='red',alpha=0.5)
```

```python
print(classification_report(y_test,y_pred_lrc))
```

# `Time Taken When Training the Model`
